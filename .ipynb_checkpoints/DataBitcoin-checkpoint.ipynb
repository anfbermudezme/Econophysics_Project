```python
%matplotlib inline
import numpy as np
import pandas as pd
import investpy
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
```

```python
data = investpy.get_crypto_historical_data(crypto='bitcoin', from_date='03/01/2010', to_date='03/01/2013')
data.head()
```

```python
data_close = data['Close']
data_close
```

```python
data_array = data_close.to_numpy()
data_array
```

```python
df = pd.DataFrame(data_array)
df
```

```python
np.savetxt('Post_Bitcoin.txt', data_array, fmt = '%g')
```

```python
fig, ax = plt.subplots()
ax.plot(df,'kx', label = "Data")
ax.set_xlabel("x")
ax.set_ylabel("y")
plt.grid()
```
